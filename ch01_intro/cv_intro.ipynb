## 1. 컴퓨터 비젼 

- (전통적인 의미) 영상 신호로 부터 상황을 인식(즉, 물체 레벨)로 인식
- 컴퓨터 비젼은 인간의 시각 시스템을 모방하면서 발전해 왔음.

- 역사 
   - 인공 지능의 영역 중 시각정보를 다루는 분야로 성장   
   - 1960~70년대	에지(Edge) 검출, 2D 선도 분석 등 기하학적 접근 중심
   - 1980년대	수학적 모델링, 광학 흐름(Optical Flow), 신경망(CNN의 시초) 연구 시작
   - 1990~2000년대	특징 추출 알고리즘(SIFT, HOG 등) 기반의 통계적 방법론 유행
   -  2010년대~현재	**딥러닝(Deep Learning)**의 도입으로 비약적인 성능 향상 (AlexNet 이후)

- 뇌는 신호처리알고리즘이 들어간 컴퓨터에 해당

###  관련 영역 

- computer vision : 영상신호로 부터 상황을 인식(즉, 물체 레벨)로 인식    
- image processing: 영상신호를 신호처리 레벨로 처리 (CV의 앞던 처리 또는 사람들이 사용)
- computer graphics: 모델로 부터 이미지/영상을 생성 (가상 현실, 게임 등) 

예전에는 각 전문적인 영역으로 연구자들이 구분되어 있었으나 2010이후에는 이 세가지 분야가 통합되어 연구자의 구분이 없어짐.
