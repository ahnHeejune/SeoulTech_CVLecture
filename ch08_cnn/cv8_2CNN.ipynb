## 2) 손실 함수
- $y_est$: 모델의 출력,  $y_{gt}$: 정답 
- MSE (mean sqaured error): 신경망 초기에는 초기에는 주로 사용  $ L(y_{gt}, y_{est}) = \frac{1}{|M|} \sum_{i \in M } (y_{gt} - y_{est})^2 $  
- L1 loss: 1차 함수 형태의 loss  $ L(y_{gt}, y_{est}) = \frac{1}{B} \sum_{i} |y_{gt} - y_{est} | $  
- Cross entropy
     - 원래는 두개의 확률 분포가 얼마나 비슷한지에 대한 척도
          - $y_{gt} = (y_{gt}[1], \cdots , y_{gt}[N])$, $ y_{est} = (y_{est}[1], \cdots , y_{est}[N])$
     - $L(y_{gt}, y_{est}) = \sum_{i} - y_{gt}[i] * log(y_{est}[i]) $
          - 0이면 동일, 클수돌 분포가 다름
          - 예, 이진 불류의 경우  $y_{gt}$는 0 또는 1이므로 확률벡터는  $[1, 0]$ 또는 $[0, 1]$ (결정적)
          - 예측 $y_{est} = [p, 1-p]$ 의 경우: 0 인경우 $-0*log(p)- 1*log(1-p)= -log(p)$
               - 즉 $p$가 작을 수록 커지고  p 가 커서 1로 갈수록 작아짐.
           

## 2) Optimizer

- 파라메터를 업데이트 하는 방법

     $$  W  = W  - \frac{\partial L}{\partial W}  = W  - \frac{1}{|M|} \sum_{M} \frac{\partial L}{\partial W} $$
- SGD (Stochastic Gradient Descent)
  - 모든 데이터를 한꺼번에 계산이 불가하므로 일부분을 가지고 학슴
    $$  W  = W  - \frac{\partial L}{\partial W}  = W  - \frac{1}{|B|} \sum_{B} \frac{\partial L}{\partial W} $$
  
  - 현재 gradient 만 사용

- 이전의 gradient를 축적하여 (관성 moment) 사용
   $$  W  = W  - \frac{\partial L}{\partial W},  V = \alpha *V  - \frac{\partial L}{ \partial W}  $$
    
   
- learning rate (LR)에 따른 차이 : 보통 0.001 정도에서 시작하여 차차 줄여 나감.

## 과정합 문제 

- 문제나 데이터에 비해 모델이 너무 크면 (파라메터가 많음) 과정합이 발생할 수 있음
- 대응 방법
    - 데이터 증강data augmentation: 훈련 집합을 조금씩 변형하여 인위적으로 늘림
    - 드롭아웃 : 특징 맵을 구성하는 요소 중 일부를 랜덤 선택하여 0으로 설정하여 학습에서 배제, 학습할 때만 적용하고 예측(추론) 과정에서는 적용 안함
    -  조기 멈춤: 성능 향상이 없으면 설정한 세대 수 이전에 학습을 멈춤
