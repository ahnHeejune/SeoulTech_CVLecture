## 1. Anaconda 또는 miniconda 사용한 가상환경 셋업

1. anacondㅁ/miniconda 설치: https://www.anaconda.com/ 
2. anaconda prompt 또는 powershell prompt 실행
3. conda create -n cvlec  python=3.10  #  cvlec 대신 원하는 환경이름을 사용,  python 버젼도 희망하는 것을 사용
4. conda activate cvlec   # 본인이 정한 환경 이름을 사용 
5. pip -V    # 환경에 맞는지 꼭 확인해야함
6. pip install opencv-python pillow  # 이외의 필요한 package는 향후 설치함. 

## 2.  VENV 사용

- Python 을 설치, venv 는 Python 자체에 포함됨

1. python -m venv cvlec   # 본인이 원하는 디렉토리위치에 가서 실행 하면 cvlev라는 디랙토리가 생길 것임. 
2. cvlec\Scripts\activate # batch 파일 실행 하면 환경 설정됨.
3. pip -V    # 환경에 맞는지 꼭 확인해야함
4. pip install opencv-python pillow  # 이외의 필요한 package는 향후 설치함. 

### Note: activate 는 뭘 하는 것인가?

- PATH 환경변수를 설정하여 해당 가상환경에 맞는 python 인터프리터와 pip 를 사용하게 함. 
- package 들은 Lib/site-packages (Windows), lib/python<version>/site-packages 아래 저장됨.



# OpenCV-Python 

OpenCV는 컴퓨터 비젼및 영상처리 연구와 실제 시스템을 개발할때 사용할 수 있는 기능을 구현해 놓은 라이브러리

- 역사 
- 지원하는 기능 영역 
- 시스템 구조 
- 기본적인 사용법

## 1. 역사 

- 1998년 Intel에서 시작 
- 2006년 Version 1 공개  (C++ 만 지원)
- 2009년 Version 2 공개  (Python, Java Binding 제공)
- 2015년 Version 3 공개  (javascript Binding 과 DNN 모둘 제공)
- 2018년 Version 4 공개 

## 관련 자료 

- 공식 홈페이지: https://opencv.org
- 매뉴얼 싸이드: https://docs.opencv.org/<version>/index.html
- 코드 (github): https://github.com/opencv/opencv
- 이외에도 수 많은 책, 블로그, 강의 동영상이 있음.

## 2. 지원 범위 

- 기존 다차원 배열 처리 (Mat, linear solver, eigen matrix, SVD etc) 
- 이미지 처리 (이미지 필터, 기하변환, 크기 변환, 컬러 변환 등)
- 고급 컴퓨터 비젼 알고리즘 (에지 검출, 영역 분할, 특징점, 매칭 등)
- 3차원 및 카메라 관련 처리 (카메라 캘리브레이션, 깊이 추정 등)
- 일부 그래픽스 처리  
- 머신 러닝/ 딥러닝 관련 기능 
- 사용자 인터페이스 (그리기, 입출력)
- 파일 입출력 (이미지, 동영상, 카메라 처리)


## 3. 구조 및 Python 연동 

-  Mat 를 처리하는 Core 모듈을 바탕으로 다른 기능들이 구성됨
-  Python은 PyBiding을 사용하여 지원하면 numpy 배열을 사용
     - 즉, Python으로 코딩하더라도내부적으로는 C++ 코드로 동작됨.
     - 내부적으로  OpenCV 의 Mat와 데이터를 호환함.
